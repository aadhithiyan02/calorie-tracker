```python
from dataclasses import dataclass
import numpy as np
import matplotlib.pyplot as plt
```

```python
calorie_goal_limit=int(input("Enter your calorie diet:"))
protein_goal=int(input("Enter Protein intake required for a day"))
fat_goal=int(input("Enter Fat intake required for a day"))
carbs_goal=int(input("Enter Carbohydrates intake required for a day"))
foods=None
```

```python
today=[]
```

```python
@dataclass
class Food:
    name:str
    calories:int
    proteins:int
    fats:int
    carbs:int
```

```python

done=False
while not done:
    print("""
    1 to Add new food intake
    2 to visualising progress
    3 to quit""")
    choice=input("Choose an option:")

    if(choice=="1"):
        print("Adding new food")
        name=input("Enter food")
        calories=int(input("enter calories:"))
        proteins=int(input("enter protein:"))
        fats=int(input("enter fat:"))
        carbs=int(input("enter carb:"))
        food=Food(name,calories,proteins,fats,carbs)
        today.append(food)
        print("Successfully added")
    elif choice=="2":
        goal=[" ","intake","goal","intake","goal","intake","goal"]
        calories_sum=sum(food.calories for food in today)
        proteins_sum=sum(food.proteins for food in today)
        fats_sum=sum(food.fats for food in today)
        carbs_sum=sum(food.carbs for food in today)
        fig,axs=plt.subplots(2,2)
        #axs[0,1].set_xticklabels(goal)
        axs[0,0].pie([proteins_sum,fats_sum,carbs_sum],labels=["Proteins","Fats","Carbs"],autopct="%1.1f%%" )
        axs[0,0].set_title("Macro nutrients distribution")
        axs[0,1].bar([0,6,12],[proteins_sum, fats_sum,carbs_sum],width=0.4)
        axs[0,1].bar([3,9,15],[protein_goal,fat_goal,carbs_goal] ,width=0.4)
        axs[0,1].set_title("Macro nutrients progress")
        axs[1,0].pie([calories_sum,calorie_goal_limit-calories_sum],labels=["calories taken","calories remaining"],autopct="%1.1f%%")
        axs[1,1].plot(list(range(len(today))), np.cumsum([foods.calories for foods in today]),label="calories eaten")
        axs[1,1].plot(list(range(len(today))),[calorie_goal_limit]*len(today),label="calories goal")
        axs[1,1].legend()
        axs[1,1].set_title("Calories Goal Vs Calories taken")
        
        fig.tight_layout()
        plt.show()
    elif choice=="3":
        done=True
        
```
